**Crowd Density Estimation using Video Analytics for
Smart City Monitoring**

This project presents a real-time crowd density monitoring system using video analytics and deep learning. We use YOLOv5, a state-of-the-art object detection model, to detect people in video frames captured from public areas. The number of people per frame is counted and classified into three density levels:

*   Low (0–10 people)
*   Medium (11–25 people)
*   High (26+ people)


The system is capable of generating live statistics such as people count vs. time and density distribution. This model can assist in monitoring public spaces, enhancing safety, and supporting smart city infrastructure.

```
import os
import matplotlib.pyplot as plt

# Path to YOLOv5 label files
labels_path = 'runs/detect/exp2/labels'  # change if your exp number is different

frame_counts = []
density_levels = []

for filename in sorted(os.listdir(labels_path)):
    if filename.endswith(".txt"):
        filepath = os.path.join(labels_path, filename)
        with open(filepath, 'r') as f:
            lines = f.readlines()
            people_count = len(lines)
            frame_counts.append(people_count)

            # Density Classification
            if people_count <= 10:
                density = 'Low'
            elif people_count <= 25:
                density = 'Medium'
            else:
                density = 'High'
            density_levels.append(density)

# Print summary
print("Sample Counts:", frame_counts[:10])
print("Sample Densities:", density_levels[:10])
```

```
import matplotlib.pyplot as plt

# X-axis: Frame number (Time)
frames = list(range(len(frame_counts)))

# Y-axis: Number of people
plt.figure(figsize=(12, 6))
plt.plot(frames, frame_counts, marker='o', linestyle='-', color='blue')
plt.title('Time vs People Count')
plt.xlabel('Frame Number')
plt.ylabel('Number of People')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
from collections import Counter

# Count how many times each density category appears
density_counts = Counter(density_levels)

# Plot as a pie chart
labels = density_counts.keys()
sizes = density_counts.values()

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['green', 'orange', 'red'])
plt.title('Crowd Density Distribution')
plt.axis('equal')  # Equal aspect ratio makes the pie circular
plt.show()
```
